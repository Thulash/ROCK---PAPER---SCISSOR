##ROCK - PAPER - SCISSOR

 
  INPUT:
  - Enter the choice : "Paper"
  - Computer generates a random key out of three

OUTPUT:
  - Based on rule we will get the winner of the round 
  - Game Board:

      - Human Points
      - Computer Points
      - Ties
      - Keep on the game until the player asks for exit

RULE OF THE GAME:
  - Rules to win the rock-paper-scissor:
      - R + R = T
      - R + P = P
      - R + S = R
      - P + P = T 
      - P + R = P 
      - P + S = S 
      - S + S = T
      - S + P = S
      - S + R = R


```
# import random functions for the bot to generate random key
import random

keys = ['P','S', 'R']
comp_point = 0
human_point = 0
tie_point = 0

rules = {"P+P":"T",
         "P+R":"P",
         "P+S":"S",
         "R+R":"T",
         "R+S":"R",
         "R+P":"P",
         "S+S":"T",
         "S+P":"S",
         "S+R":"R"
         }

choice = ' '
while choice!= ' E ':
  # display a menu for choice
  print("\nMENU!")
  print("P - Paper")
  print("S - Scissor")
  print("R - Rock")
  print("E - Exit Game!")

  #input the choice from Human
  choice = str(input("\nEnter your choice: "))
  choice = choice.upper()

  print("\nHuman Choice: ", choice)

   
  #check the choice validity
  if choice in keys  :
    pass
  elif choice == 'E' or choice =='e':
    print("\nEND GAME!")
    break
  else :
    print("\nInvalid choice!","\nTRY AGAIN!!!")
    continue

  #generate random key for the round
  bot = random.choice(keys)
  print("Computer choice: ", bot)
  
  # finding the winner of the round
  res = choice + "+" + bot
  winner = rules.get(res)
  
  if winner == choice:
    print("\nHuman wins!")
    human_point = human_point + 1
  elif winner == bot :
    print("\nComputer Wins!")
    comp_point = comp_point + 1
  else :
    print("\nGame is Tie!")
    tie_point = tie_point + 1

  # gameboard
  print("\nGAMEBOARD\n")
  print("Human Points: ", human_point)
  print("Computer Points: ", comp_point)
  print("Tie Points: ", tie_point)
```
